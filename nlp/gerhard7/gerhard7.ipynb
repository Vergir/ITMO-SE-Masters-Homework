```
import pandas
import pathlib
import flair
import flair.data_fetcher
import flair.visual.training_curves
import matplotlib
import pylab
```

```
data = pandas.read_csv("https://raw.githubusercontent.com/gwohlgen/misc/master/spam.csv", encoding='latin-1').drop_duplicates()

data = data[['v1', 'v2']].rename(columns={"v1":"label", "v2":"text"})
data['label'] = '__label__' + data['label'].astype(str)

data.iloc[0:int(len(data)*0.6)].to_csv('train.csv', sep='\t', index = False, header = False)
data.iloc[int(len(data)*0.6):int(len(data)*0.8)].to_csv('test.csv', sep='\t', index = False, header = False)
data.iloc[int(len(data)*0.8):].to_csv('dev.csv', sep='\t', index = False, header = False);


print(len(data))
data.head()
```

```
DOWNSAMPLE_RATE = 0.1
MAX_EPOCHS = 5

embeddings_list = [
    (flair.embeddings.CharacterEmbeddings(), 'Character Embeddings'),
    (flair.embeddings.BytePairEmbeddings('en'), 'BytePair Embeddings'),
    (flair.embeddings.WordEmbeddings('glove'), 'GloVe'), 
    (flair.embeddings.WordEmbeddings('extvec'), 'ExtVec'),
    (flair.embeddings.FlairEmbeddings('news-forward'), 'fastText Embeddings from News'),
    (flair.embeddings.FlairEmbeddings('mix-forward'), 'fastText Embeddings from mixed sources')
]

for embeddings_data in embeddings_list:
  #load things
  corpus = flair.data_fetcher.NLPTaskDataFetcher.load_classification_corpus(
    pathlib.Path('./'), 
    test_file='test.csv', 
    dev_file='dev.csv', 
    train_file='train.csv'
  ).downsample(DOWNSAMPLE_RATE)
  embeddings = flair.embeddings.DocumentPoolEmbeddings([embeddings_data[0]])
  classifier = flair.models.TextClassifier(
    embeddings, 
    label_dictionary=corpus.make_label_dictionary(), 
    multi_label=False
  )
  
  # train
  trainer = flair.trainers.ModelTrainer(classifier, corpus)
  trainer.train('./', max_epochs=MAX_EPOCHS)
  
  
  # show stats
  print('--------------------')
  print('--------------------')
  print(embeddings_data[1])
  plotter.plot_training_curves('loss.tsv')
  %pylab inline
  pylab.rcParams['figure.figsize'] = 25, 25
  image = matplotlib.image.imread('training.png')
  matplotlib.pyplot.imshow(image)
  matplotlib.pyplot.show()
```
